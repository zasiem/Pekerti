```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import scipy

#read csv
data = pd.read_csv('pekerti.csv')
data.head()
```

```python
def silhouette_score(syarat1, syarat2, syarat3, syarat4) :
    data_selection = data[[syarat1,syarat2,syarat3,syarat4]]
    
    cluster = KMeans(n_clusters=2).fit(data_selection)
    cluster.labels_
    cluster_label = cluster.predict(data_selection)
    
    centroids = cluster.cluster_centers_
  
    score = silhouette_score(data, cluster_label, metric='euclidean')
    
    return score
```

```python
#techno
techno = silhouette_score('OOP','RPB','RPL','PI')
techno
```
