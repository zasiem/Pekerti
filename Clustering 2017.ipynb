```python
matriks = pd.DataFrame(data.NIM)
matriks['Technopreneur'] = techno.Label
matriks['EAD'] = ead.Label
matriks['ERP'] = erp.Label
matriks['EDM'] = edm.Label
matriks['EA'] = ea.Label
matriks['ISM'] = ism.Label
matriks['EIM'] = eim.Label
matriks = matriks.sort_values(by = 'NIM').iloc[22::]
matriks.head()
```

```python
# matriks = matriks.sort_values(by = 'NIM').iloc[22::]
```

```python
hasilPeminatan = pd.read_excel('hasilpeminatanpipe.xlsx')
hasilPeminatan = hasilPeminatan.drop(['No','Nama','Mata Kuliah','Kode Mata Kuliah','Kelas Mata Kuliah','ID'],axis=1)
hasilPeminatan = hasilPeminatan.sort_values(by ='NIM').iloc[11::]
hasilPeminatan.head()
```

```python
# hasilPeminatan = hasilPeminatan.sort_values(by ='NIM')
# hasilPeminatan.iloc[11::]
# hasilPeminatan
```

```python
temp = [0] * len(matriks)
for indeks, baris in matriks.iterrows() :
    for index, row in hasilPeminatan.iterrows() :
        if baris['NIM'] == row['NIM'] :
            if row['Peminatan'] == "Information System Management" :
                temp[indeks] = "ISM"
            elif row['Peminatan'] == "Enterprise System Development" :
                temp[indeks] = "EAD"
            elif row['Peminatan'] == "Enterprise Data Management" :
                temp[indeks] = "EDM"
            elif row['Peminatan'] == "Enterprise Architecture" :
                temp[indeks] = "EA"
            elif row['Peminatan'] == "Technopreneur" :
                temp[indeks] = "TECHNO"
            elif row['Peminatan'] == "Enterprise Infrastructure Management" :
                temp[indeks] = "EIM"
            elif row['Peminatan'] == "Enterprise Resource Planning" :
                temp[indeks] = "ERP"
            break
```

```python
matriks["Peminatan"] = temp
matriks.head()
```

```python
isCocok = [0] * len(matriks)
for index, row in matriks.iterrows():
    if row['Peminatan'] == 'TECHNO':
        if row['Technopreneur'] == 1 :
            isCocok[index] = "Cocok"
        else :
            isCocok[index] = "Tidak Cocok"
    elif row['Peminatan'] == 'EAD':
        if row['EAD'] == 1 :
            isCocok[index] = "Cocok"
        else :
            isCocok[index] = "Tidak Cocok"
    elif row['Peminatan'] == 'EDM':
        if row['EDM'] == 1 :
            isCocok[index] = "Cocok"
        else :
            isCocok[index] = "Tidak Cocok"
    elif row['Peminatan'] == 'ERP':
        if row['ERP'] == 1 :
            isCocok[index] = "Cocok"
        else :
            isCocok[index] = "Tidak Cocok"
    elif row['Peminatan'] == 'EAA' or row['Peminatan'] == 'EA':
        if row['EA'] == 1 :
            isCocok[index] = "Cocok"
        else :
            isCocok[index] = "Tidak Cocok"
    elif row['Peminatan'] == 'ISM':
        if row['ISM'] == 1 :
            isCocok[index] = "Cocok"
        else :
            isCocok[index] = "Tidak Cocok"
    elif row['Peminatan'] == 'EIM':
        if row['EIM'] == 1 :
            isCocok[index] = "Cocok"
        else :
            isCocok[index] = "Tidak Cocok"
```

```python
matriks["Keterangan"] = isCocok
matriks.head()
```

```python
jumlah_cocok = 0
jumlah_tidak_cocok = 0
jumlah_tidak_memilih = 0

for index, row in matriks.iterrows():
    if row['Keterangan'] == "Cocok" :
        jumlah_cocok = jumlah_cocok + 1
    elif row['Keterangan'] == "Tidak Cocok" :
        jumlah_tidak_cocok = jumlah_tidak_cocok + 1
    else :
        jumlah_tidak_memilih = jumlah_tidak_memilih + 1
   
        
data_cocok = [['Cocok', jumlah_cocok],
              ['Tidak Cocok', jumlah_tidak_cocok],
              ['Total Mahasiswa', len(matriks)]] 

total_cocok = pd.DataFrame(data_cocok, columns = ['Status', 'Jumlah']) 
  
total_cocok 
```

```python
labels = 'Cocok', 'Tidak Cocok'
sizes = [jumlah_cocok, jumlah_tidak_cocok]
colors = ['green', 'red']
explode = (0.1, 0)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
