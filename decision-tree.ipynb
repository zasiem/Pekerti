```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read master data csv
masterData = pd.read_csv("masterdata2016.csv")
masterData.head()
```

```python
#read data peminatan
dataPeminatan = pd.read_csv("2016 data v5.csv")
dataPeminatan.head()
```

```python
#merge table
masterData["Pilihan 1"] = dataPeminatan['Pilihan 1']
masterData["Pilihan 2"] = dataPeminatan['Pilihan 2']
masterData["Peminatan"] = dataPeminatan['Peminatan']
masterData.head()
```

```python
#get skor 1
skor1 = [0] * len(dataPeminatan)
for index, row in masterData.iterrows():
    if row["Pilihan 1"] == "TECHNO" :
        skor1[index] = row["OOP"] + row["RPB"] + row["WEB"] + row["PI"] + 1
    elif row['Pilihan 1'] == "EAD" :
        skor1[index] = row["OOP"] + row["APSI"] + row["RPB"] + row["WEB"] + 1
    elif row['Pilihan 1'] == "EDM" :
        skor1[index] = row["Statistik"] + row["Matdis"] + row["Alpro"] + row["Strukdat"] + 1
    elif row['Pilihan 1'] == "ERP" :
        skor1[index] = row["SE"] + row["RPB"] + row["PO"] + row["SCM"] + 1
    elif row['Pilihan 1'] == "EA" :
        skor1[index] = row["RPB"] + row["APSI"] + row["EA"] + row["Basdat"] + 1
    elif row['Pilihan 1'] == "EIM" :
        skor1[index] = row["Desjar"] + row["Manjarkom"] + row["Sisop"] + row["Manprosi"] + 1
    elif row['Pilihan 1'] == "ISM" :
        skor1[index] = row["SE"] + row["Manjarkom"] + row["Sisop"] + row["MSDM"] + 1

len(skor1)
```

```python
#get skor 2
skor2 = [0] * len(dataPeminatan)
for index, row in masterData.iterrows():
    if row["Pilihan 2"] == "TECHNO" :
        skor2[index] = row["OOP"] + row["RPB"] + row["WEB"] + row["PI"]
    elif row['Pilihan 2'] == "EAD" :
        skor2[index] = row["OOP"] + row["APSI"] + row["RPB"] + row["WEB"]
    elif row['Pilihan 2'] == "EDM" :
        skor2[index] = row["Statistik"] + row["Matdis"] + row["Alpro"] + row["Strukdat"]
    elif row['Pilihan 2'] == "ERP" :
        skor2[index] = row["SE"] + row["RPB"] + row["PO"] + row["SCM"]
    elif row['Pilihan 2'] == "EA" :
        skor2[index] = row["RPB"] + row["APSI"] + row["EA"] + row["Basdat"]
    elif row['Pilihan 2'] == "EIM" :
        skor2[index] = row["Desjar"] + row["Manjarkom"] + row["Sisop"] + row["Manprosi"]
    elif row['Pilihan 2'] == "ISM" :
        skor2[index] = row["SE"] + row["Manjarkom"] + row["Sisop"] + row["MSDM"]
        
len(skor2)
```

```python
if "Skor 1" in masterData.columns :
    del masterData['Skor 1']
if "Skor 2" in masterData.columns :
    del masterData['Skor 2']
    
masterData['Skor 1'] = skor1
masterData['Skor 2'] = skor2

#clean null tuples
masterData = masterData.dropna(how='any',axis=0) 
cleanData = masterData[["No","NIM","Skor 1","Pilihan 1","Skor 2","Pilihan 2","Peminatan"]]
cleanData.head()
```
