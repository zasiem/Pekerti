```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#read csv
data = pd.read_csv('pekerti.csv')
data.head()
```

```python
New_data = data[['RPB','APSI','EA','Basdat']]
New_data.head()
```

```python
ratarata = pd.DataFrame(data['NIM'])
ratarata['Ratarata'] = New_data.mean(axis=1)
ratarata.head()
```

```python
ratarata.insert(loc=0, column='No', value=np.arange(start=1,stop=len(data)+1,step=1))
ratarata.head()
```

```python
#seleksi 2016 aja dan tentukan kmeansnya
new_ratarata_ea = ratarata.iloc[:403]
new_ratarata_ea.iloc[:10]
```

```python
new_ratarata_ea.plot(kind="scatter",x="No",y="Ratarata")
```

```python
selection = new_ratarata.iloc[:,::2]
cluster = KMeans(n_clusters=2).fit_predict(selection)
cluster
```
