```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#read csv
data = pd.read_csv('2016 data filtered.csv')
NewData = data.drop(['Pilihan 1','Pilihan 2', 'Peminatan', 'No','NIM'], axis=1)
NewData.head()
```

```python
def getScore(syarat1, syarat2, syarat3, syarat4) :
    data_selection = data[[syarat1,syarat2,syarat3,syarat4]]
    
    cluster = KMeans(n_clusters=2).fit(data_selection)
    cluster.labels_
    cluster_label = cluster.predict(data_selection)
  
    score = silhouette_score(data_selection, cluster_label, metric='euclidean')
    
    return score
```

```python
#get all shilhouette score
techno = getScore('OOP','RPB','RPL','PI')
ead = getScore('OOP','APSI','RPB','WEB')
edm = getScore('Statistik', 'Matdis', 'Alpro', 'Strukdat')
erp = getScore('SE','RPB','PO','SCM')
ea = getScore('RPB','APSI','EA','Basdat')
eim = getScore('Desjar','Manjarkom','Sisop','KSI')
ism = getScore('SE','Manjarkom','Sisop','MSDM')
```

```python
outputData = [
    ['Techno',techno],
    ['EAD',ead],
    ['EDM',edm],
    ['ERP',erp],
    ['EA',ea],
    ['EIM',eim],
    ['ISM',ism]
]

output = pd.DataFrame(
    outputData,
    columns = ['Peminatan', 'Shilouette Score']
)

output
```
